```python
import numpy as np
import sys

#Questo serve per visualizzare senza accorciamenti e sintesi i printaggi di numpy
np.set_printoptions(threshold=sys.maxsize, linewidth=250)

#---------------------------------------------------------

#Voglio leggere un file generico TXT con al suo interno una matrice divisa per linee da \n e spazi tra gli elementi
with open("MATRIX.txt", "r") as file_text_1:
    content_matrix = file_text_1.read()

#Voglio leggere un file generico TXT con al suo interno una lista di parole divise da \n
with open("LIST_WORDS.txt", "r") as file_text_2:
    content_words = file_text_2.read()

#---------------------------------------------------------

#Spacco il testo in parole utilizzando \n come separatore
words = content_words.split("\n")

#Spacco il testo in linee utilizzando \n come separatore
lines = content_matrix.split("\n")

#Determino il numero delle colonne e righe
num_rows = len(lines)
num_cols = len(lines[0].split())

#Creo una matrice vuota di numpy delle dimensioni trovate prima di tipologia string
matrix = np.empty((num_rows, num_cols), dtype=str)

#Aggiungo gli elementi delle linee di prima all'interno della matrice
for i, line in enumerate(lines):
    matrix[i] = line.split()

#---------------------------------------------------------

#Le variabili sono matrix e words
```

```python
def findHorizontalWord(matrix, words):
    '''Questa funzione trova le parole orrizzontali, anche scritte al contrario, all'interno della matrice con "n" righe per "m" colonne in cui "n" e "m" sono numeri maggiori o uguale a 2,
    'matrix' è una matrice di numpy ed è composta da singole lettere maiuscole,
    'words' è una lista di stringhe scritte in maiuscolo che contiene le parole da trovare,  
    quando la parola per intero è trovata le lettere corrispondenti nella matrice verranno sostituite dal simbolo "_" 
    e la matrice sarà salvata alla fine, inoltre verrà restituita una lista con le parole trovate'''
    words_found = []
    #Ciclo per ogni parola nella lista di parole
    for word in words:
        #Ciclo per ogni riga della matrice
        for i in range(len(matrix)):
            #Ciclo per ogni elemento della riga della matrice
            for j in range(len(matrix[i]) - len(word) + 1):
                #Se la parola è uguale alla porzione di riga della matrice allora la parola è stata trovata
                if word == "".join(matrix[i][j:j+len(word)]):
                    #Aggiungo la parola alla lista delle parole trovate
                    words_found.append(word)
                    #Sostituisco le lettere della parola trovata con il simbolo "_"
                    matrix[i][j:j+len(word)] = ["_"] * len(word)
                #Se la parola è uguale alla porzione di riga della matrice al contrario allora la parola è stata trovata    
                elif word == "".join(matrix[i][j:j+len(word)][::-1]):
                    #Aggiungo la parola alla lista delle parole trovate
                    words_found.append(word)
                    #Sostituisco le lettere della parola trovata con il simbolo "_"
                    matrix[i][j:j+len(word)] = ["_"] * len(word)
    return words_found

def findVerticalWord(matrix, words):
    '''Questa funzione trova le parole verticali, anche scritte al contrario, all'interno della matrice con "n" righe per "m" colonne in cui "n" e "m" sono numeri maggiori o uguale a 2,
    'matrix' è una matrice di numpy ed è composta da singole lettere maiuscole,
    'words' è una lista di stringhe scritte in maiuscolo che contiene le parole da trovare,  
    quando la parola per intero è trovata le lettere corrispondenti nella matrice verranno sostituite dal simbolo "_" 
    e la matrice sarà salvata alla fine, inoltre verrà restituita una lista con le parole trovate'''
    words_found = []
    #Ciclo per ogni parola nella lista di parole
    for word in words:
        #Ciclo per ogni riga della matrice
        for i in range(len(matrix) - len(word) + 1):
            #Ciclo per ogni elemento della riga della matrice
            for j in range(len(matrix[i])):
                #Creo una lista di lettere che contiene la porzione di colonna della matrice
                column = [matrix[k][j] for k in range(i, i+len(word))]
                #Se la parola è uguale alla porzione di colonna della matrice allora la parola è stata trovata
                if word == "".join(column):
                    #Aggiungo la parola alla lista delle parole trovate
                    words_found.append(word)
                    #Sostituisco le lettere della parola trovata con il simbolo "_"
                    for k in range(i, i+len(word)):
                        matrix[k][j] = "_"
                #Se la parola è uguale alla porzione di colonna della matrice al contrario allora la parola è stata trovata    
                elif word == "".join(column[::-1]):
                    #Aggiungo la parola alla lista delle parole trovate
                    words_found.append(word)
                    #Sostituisco le lettere della parola trovata con il simbolo "_"
                    for k in range(i, i+len(word)):
                        matrix[k][j] = "_"
    return words_found

def findDiagonalWord(matrix, words):
    '''Questa funzione trova le parole diagonali, anche scritte al contrario, all'interno della matrice con "n" righe per "m" colonne in cui "n" e "m" sono numeri maggiori o uguale a 2,
    'matrix' è una matrice di numpy ed è composta da singole lettere maiuscole,
    'words' è una lista di stringhe scritte in maiuscolo che contiene le parole da trovare,  
    quando la parola per intero è trovata le lettere corrispondenti nella matrice verranno sostituite dal simbolo "_" 
    e la matrice sarà salvata alla fine, inoltre verrà restituita una lista con le parole trovate'''
    words_found = []
    #Ciclo per ogni parola nella lista di parole
    for word in words:
        #Ciclo per ogni riga della matrice
        for i in range(len(matrix) - len(word) + 1):
            #Ciclo per ogni elemento della riga della matrice
            for j in range(len(matrix[i]) - len(word) + 1):
                #Creo una lista di lettere che contiene la porzione di diagonale della matrice
                diagonal = [matrix[i+k][j+k] for k in range(len(word))]
                #Se la parola è uguale alla porzione di diagonale della matrice allora la parola è stata trovata
                if word == "".join(diagonal):
                    #Aggiungo la parola alla lista delle parole trovate
                    words_found.append(word)
                    #Sostituisco le lettere della parola trovata con il simbolo "_"
                    for k in range(len(word)):
                        matrix[i+k][j+k] = "_"
                #Se la parola è uguale alla porzione di diagonale della matrice al contrario allora la parola è stata trovata    
                elif word == "".join(diagonal[::-1]):
                    #Aggiungo la parola alla lista delle parole trovate
                    words_found.append(word)
                    #Sostituisco le lettere della parola trovata con il simbolo "_"
                    for k in range(len(word)):
                        matrix[i+k][j+k] = "_"
    return words_found
                
```

```python
#Questa lista contiene le parole orrizzontali trovate
h_word = findHorizontalWord(matrix, words)
#Questa lista contiene le parole verticali trovate
v_word = findVerticalWord(matrix, words)
#Questa lista contiene le parole diagonali trovate
d_word = findDiagonalWord(matrix, words)
```
